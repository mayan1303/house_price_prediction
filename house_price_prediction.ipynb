```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing

# data = fetch_california_housing()
# X = data.data
# y = data.target
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df=fetch_california_housing()
df
```

```python
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
# Independent and dependent features
X=dataset
y=df.target
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)#30% data -> testing 70% data->training
```

```python
X_train
```

```python
#standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)#learn+apply
X_train
""" when we use fit f
it() learns statistics from data:
mean
standard deviation

For StandardScaler:
z = (x - mean) / std


                     """
```

```python
X_test=scaler.transform(X_test)#only apply
```

```python
scaler.inverse_transform(X_train)
```

```python
#implement linear regression
from sklearn.linear_model import LinearRegression
"""cross validation- internally splits train datsets into
multiple train test"""
"""every cross validation has different accuracy 
so at last we find mean of that and give result"""

from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
##prediction
reg_predict=regression.predict(X_test)
```

```python
reg_predict
```

```python
import seaborn as sns
sns.displot(reg_predict-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_predict)
score
```
